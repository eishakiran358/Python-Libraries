# What is Pandas?
Pandas is an open-source Python library used for data manipulation, data analysis, and data cleaning.

It provides fast, flexible, and easy-to-use data structures such as:

- Series (1D)
- DataFrame (2D)

Pandas is one of the most important libraries for Data Science, Machine Learning, Artificial Intelligence, and Data Analysis.

Pandas helps us:

- Read datasets
- Clean data

- Filter data
- Sort data
- Handle missing values
- Perform statistical analysis
- Prepare data for Machine Learning

# Handling Missing Values in Pandas

Missing values are empty or unavailable values in a dataset.

In Pandas, missing values are usually represented as:

- NaN
- None
- NaT

Titanic dataset mein important missing columns:

- age
- deck
- embark_town

```python
# 1. Check Missing Values
df.isnull()
```

```python

df.isnull().sum()
# True  = Missing Value
# False = Value Available
```

```python
# 2. Missing Values Percentage ⭐
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100
missing_percentage
```

```python
# 2nd way 
missing_data = pd.DataFrame({
    "Missing Values": df.isnull().sum(),
    "Percentage": (df.isnull().sum() / len(df)) * 100
})

missing_data
```

```python
# 3. Check Specific Column
df["age"].isnull().sum()
df["deck"].isnull().sum()
df["embark_town"].isnull().sum()
```

```python
# Select Rows with Missing Age
df[df["age"].isnull()]
```

```python
# dropna() — Remove Missing Rows
# Remove rows where ANY value is missing:
df.dropna()
```

```python
# Remove rows only if Age is missing:
df.dropna(subset=["age"])
```

```python
#Remove rows if Age OR Embark Town is missing:
df.dropna(subset=["age", "embark_town"])
```

```python
# 7. fillna() — Fill Missing Values
# Fill Age with Mean
mean_age = df["age"].mean()

df["age"] = df["age"].fillna(mean_age)
```

```python
#Fill Age with Median 
median_age = df["age"].median()

df["age"] = df["age"].fillna(median_age)
```

```python
# 8. ffill() — Forward Fill

# Previous value se missing value fill hoti hai.

df["age"] = df["age"].ffill()
```

```python
# 9. bfill() — Backward Fill

# Next available value se missing value fill hoti hai.

df["age"] = df["age"].bfill()
```

```python
# interpolate()
# Values ke beech estimated value fill karta hai.

df["age"] = df["age"].interpolate()
```

# GroupBy and Aggregation in Pandas

groupby() is used to divide data into groups.
Aggregation functions are used to calculate:

- count
- sum
- mean
- median
- minimum
- maximum
- standard deviation

```python
# Group by Sex
df.groupby("sex")
# Usually hum aggregation ke saath use karte hain:
df.groupby("sex")["survived"].mean()
```

```python
# Count Passengers by Class
df.groupby("class").size()
# or 2nd methoddddddddddd
df["class"].value_counts()
```

```python
# Average Age by Class
df.groupby("class")["age"].mean()
```

```python
# Average Fare by Class
df.groupby("class")["fare"].mean()
```

```python
#Multiple Aggregations
df.groupby("class")["fare"].agg([
    "count",
    "mean",
    "median",
    "min",
    "max"
])
```

```python
# Different Aggregations on Different Columns ⭐⭐⭐
df.groupby("class").agg({
    "age": "mean",
    "fare": "mean",
    "survived": "mean"
})
```

```python
# GroupBy + Multiple Aggregations
df.groupby(["sex", "class"]).agg({
    "survived": ["count", "mean"],
    "age": "mean",
    "fare": "mean"
})
```
